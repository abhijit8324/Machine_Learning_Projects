```python
# fibbonaci series

import time

def fib(n):
  if n == 1 or n ==0:
    return 1
  else:
    return fib(n-1) + fib(n-2)

start= time.time()
print(fib(5));
print(time.time() - start)
```

```python
import time

def fib(n,d):
  if n in d:                            # check if results already exists in the dictionary
    return d[n]
  else:
    d[n] = fib(n-1,d) + fib(n-2, d)     # compute & store in dictionary
    return d[n]

# Intializing base case in the dictionary

d = {0: 1, 1: 1}

# Timing the execution

start_time = time.time()
print(fib(50, d))       # Testing with larger number
print(f"Time taken : {time.time() - start_time} seconds")
```
