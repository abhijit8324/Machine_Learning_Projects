# Variable- Quantitative and Qualitative

# Quantitive Variable- data that is represented in numbers used for calculation.
Eg- height, weight, age

# Qualiative variable- data in grops or categories
eg- color, gender, grade


# Quantitaive Variable


1.   Discrete Variable - measured in whole number { # no of dogs, cat people- 1,2 3, ...}
2.   Continuous Variable - measured in any numerical value {# blood pressure, heart rate, height..}




# Qualitative Variable


1.   Ordinal variable - that has logical orders { exam grades->A+, A, B+..}
2.   Nominal variable - no logical orders { gender, blood group, color}

# central tendency: - Best middle value

*   Mean -  Average value
*   Median - Middle value
*   Mode - frequent values

Outliers- extreme data values
datapoint(dp's) that differe significantly from other datapoint(dp's)
dp's should be always measurable, can be found only in quantitative data

Outliers can also be negative value





datapoint - each and every row is called datapoint

example:
dp's ->{1,2,3,4,5,5,5,6,8,9,30,40}

* mean -> Average = {addition of all dp's / number of dp's}

* median -> middle value = {split data in 2 equal half, then the middle one is median} - 5

* mode -> most frequent repetead value {5}

** mean is influenced by the outliers but median is not, so median is used to replace missing values**

**mean approximately equal to median, it means no outliers in the data**

**mean and median is used for numerical data but mode is never a parameter for numerical data**

**mode is only used for categorical data**

# Mean, Median --> Numerical data
# Mode ----------> categorical data

example for mode:
color-

yellow - 20 balls
blue   - 15 balls
Red    - 10 balls
Green  - 40 balls

Then **Mode** for above will be **Green**
so if we want to use missing value of categorical in nature, then we use **Mode**

above model is **Uni-model** where we have only 1 mode

example 2 model

S- Shirt - 15
M- Shirt - 8
L- Shirt - 15
XL- Shirt - 10

Mode is S, L (most repetative value)
this model is called **Bi-model**
