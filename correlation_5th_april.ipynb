```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = [10, 8, 2, 1, 5, 6]
s = [2, 3, 9, 7, 6, 5]

data = pd.DataFrame({'Anxiety': a, 'Score': s}, index = list(range(0,6)))
data
```

```python
# scatter plot

sns.scatterplot(x = 'Anxiety', y = 'Score', data=data)
plt.show()
```

```python
# heat map is used to represent correlation

sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
num_data = titanic.select_dtypes(include=['int', 'float'])
plt.figure(figsize=(10,9))
sns.heatmap(num_data.corr(), annot=True)
plt.show()
```

```python

plt.figure(figsize=(10,9))
sns.heatmap(num_data.corr(), annot=True, cmap='Purples')
plt.show()
```

```python
from matplotlib import colormaps
list(colormaps)
```

```python
# to trim upper part

mask = np.triu(num_data.corr())
plt.figure(figsize=(10,9))
sns.heatmap(num_data.corr(), annot=True, cmap='Purples', mask=mask)
plt.show()
```

```python
# to trim lower part

mask = np.tril(num_data.corr())
plt.figure(figsize=(10,9))
sns.heatmap(num_data.corr(), annot=True, cmap='Purples', mask=mask)
plt.show()
```

```python
mask = np.triu(num_data.corr())
plt.figure(figsize=(10,9))
sns.heatmap(num_data.corr(), annot=True, cmap='Purples', fmt ='.2f', mask=mask)
plt.show()
```

```python
penguin = sns.load_dataset('penguins')
penguin.head()
```

```python
num_pg_data = penguin.select_dtypes(include=['int', 'float'])
sns.heatmap(num_pg_data.corr(), annot=True, fmt='.2f', mask=np.triu(num_pg_data.corr()), cmap='Blues')
plt.show()
```
