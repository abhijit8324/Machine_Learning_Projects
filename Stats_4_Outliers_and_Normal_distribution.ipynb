```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data = [1,2,3,4,5,5,5,6,8,9,16,17,23,24,28,65,70,200,300]
```

```python
df = pd.DataFrame(data, columns=['input'])
df
```

```python
# Stats discription/summary

df.describe()
```

```python
# distribution

sns.distplot(df['input'])
```

```python
sns.displot(df['input'])
```

```python
# distribution
sns.displot(df['input'], kind='kde')
plt.show()
```

```python
# boxplot shows outliers

sns.boxplot(df['input'], orient='h')
```

```python
# to check limit for upper fence and lower fence

# need to calculate IQR

Q1 = df['input'].quantile(0.25)
Q3 = df['input'].quantile(0.75)

IQR = Q3-Q1

lower_fence = Q1-1.5*(IQR)
upper_fence = Q3+1.5*(IQR)
```

```python
print(lower_fence, upper_fence, sep='\n')
```

```python
# x < lowerfence or x> upperfence are the outliers
```

```python
# capping the outliers to fence value

df['input'] = np.where(df['input']>upper_fence, upper_fence, df['input'])
```

```python
sns.boxplot(df.input, orient='h')
```

```python
# numerical summary

df.describe()
```

```python
df
```

```python
df['treated'] = np.where(df['input']>upper_fence, upper_fence, df['input'])
```

```python
sns.boxplot(df.treated, orient='h')
```

```python
df.describe
df
```

```python
sns.displot(df.treated, kind='kde')
plt.show()
```

```python
# capping the outliers to fence value

df['removed'] = np.where(df['input']>upper_fence, np.nan, df['input'])
df
```

```python
sns.boxplot(df.removed, orient='h')
```

```python
df.describe()
```

```python
sns.displot(df.removed, kind='kde')
```
