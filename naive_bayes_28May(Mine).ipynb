### 1. the roc_auc_score represents the ability of model to distinguish b/w positive and negative classes.
### 2. also roc_curve is used to determine the right threshold for distinguishing b/w the different classes.
### 3. AUC is the area under the roc_curve

## example 1: cancer detection model

#### 1. false positive: If in reality person didnt have cancer but our model said they will have cancer.
#### 2. false negative: reality --> cancer, but model predicted --> no cancer.

### ** false negative will be a big issue**

## example 2: Finance domain(credit card risk model)

### -- my prediction --> whether the person will default on loan or not?

#### 1. false positive : reality not default --> but our model says they will default.
#### 2. false negative : they default but model says they wont default

### thus, the threshold of 0.5 is not always the correct threshold.

### but, what is right threshold then?

#### answer is AUC-ROC curve. internally take a lot of thresholds and plot a graph b/w TPR and FPR, for whichever threshold, the AUC is max, it will give that AUC.

## TPR = TP/TP+FN = fraction of correct positive predictions from the actual positives.

FPR = FP/FP+TN ==> fraction of actual negative classes which were predicted wrongly.
