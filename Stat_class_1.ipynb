# Simpson's paradox- above middle graph is pattern for simpson paradox

to check if the grpah has simpson paradox or not we can use "hue"

```python
sns.pairplot(data, x_vars=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm'], y_vars=['body_mass_g'], kind='reg', height=5,aspect=1, hue='species')
```
