```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data from sns
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
# 5-number summary for the entire data

data.describe()
```

```python
for cols in data.columns:
    if data[cols].dtype != '0':
        sns.boxplot(data[cols], orient='h')
        plt.show()
```

```python
print('Min:', data['body_mass_g'].min())
print('Q1:', data['body_mass_g'].quantile(0.25))
print('Q2:', data['body_mass_g'].quantile(0.50))
print('Median:', data['body_mass_g'].median())
print('Q3:', data['body_mass_g'].quantile(0.75))
print('max:', data['body_mass_g'].max())
```
