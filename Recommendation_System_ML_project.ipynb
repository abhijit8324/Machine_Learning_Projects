```python
## ['Action', 'Adventure', 'Fantasy', 'Science Fiction']

import ast

ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# Cleaning and storing the data

def context(obj):
  L = []
  for i in ast.literal_eval(obj):
      L.append(i['name'])
  return L
```

```python
movies['genres']. apply(context)
```

```python
movies['genres'] = movies['genres'].apply(context)
```

```python
# explanation of ast

obj = '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}]'
print(type(obj))
new_obj = ast.literal_eval(obj)
print(type(new_obj))
L =[]
for i in new_obj:
  L.append(i['name'])
print(L)
```

```python
# Keywords

movies.iloc[0].keywords
```

We have to again extract the name, so we will use the same helper function

```python
movies['keywords'] = movies['keywords'].apply(context)
```

```python
movies
```

```python
# cast

movies.iloc[0].cast
```

```python
# we need to filter the cast name

def context_cast(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter < 3:
      L.append(i["name"])
      counter +=1
    else:
      break
  return L
```

```python
movies['cast'] = movies['cast'].apply(context_cast)
```

```python
movies
```

# next we will try to separate director in the crew column

```python
movies.iloc[0].crew
```

```python
# we will separate director from crew column by checking the job as director

def context_crew(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
  return L
```

```python
movies['crew'] = movies['crew'].apply(context_crew)
```

```python
movies
```

```python
# need to remove the spaces between words

def collaspe(obj):
  L = []
  for i in obj:
    L.append(i.replace(" ", ""))
  return L
```

```python
movies['cast'] = movies['cast'].apply(collaspe)
movies['crew'] = movies['crew'].apply(collaspe)
movies['genres'] = movies['genres'].apply(collaspe)
movies['keywords'] = movies['keywords'].apply(collaspe)
```

```python
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies
```

```python
movies['tag'] = movies['overview'] + movies['cast'] + movies['crew'] + movies['genres'] + movies['keywords']
```

```python
new_movie_data = movies.drop(columns=['cast', 'overview' , 'crew', 'genres', 'keywords'])
```

```python
new_movie_data
```

```python
new_movie_data['tag'] = new_movie_data['tag'].apply(lambda x : " ".join(x))
```

#### we will using word to vector(embedding)

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000)
```

```python
vector = cv.fit_transform(new_movie_data['tag']).toarray()
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
  index = new_movie_data[new_movie_data['title'] == movie].index[0]
  print(index)
  # The arguments 'reverse' and 'key' should be passed to 'sorted', not 'list'
  distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
  for i in distances[1:6]:
    print(new_movie_data.iloc[i[0]].title)
    pass
```
